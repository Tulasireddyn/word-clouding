```python
import requests
from bs4 import BeautifulSoup

# Example URL to scrape
url = 'https://woxsen.edu.in/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Extracting the text from the webpage
text = soup.get_text()

# Display a small portion of the text
print(text[:1000])  # Showing a snippet for demonstration purposes
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download the 'punkt_tab' data
nltk.download('punkt_tab') # Added this line to download the punkt_tab data

nltk.download('punkt')
nltk.download('stopwords')

# Clean and preprocess the text
def preprocess_text(text):
    # Lowercasing the text
    text = text.lower()

    # Removing non-alphabetic characters
    text = re.sub(r'[^a-z\s]', '', text)

    # Tokenizing
    words = word_tokenize(text)

    # Removing stop words
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]

    return words

# Preprocessing the scraped text
cleaned_words = preprocess_text(text)

# Displaying a snippet of cleaned words
print(cleaned_words[:50])
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Creating a word cloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    max_words=150,
    colormap='plasma',
    contour_color='black',
    contour_width=1,
    prefer_horizontal=0.8,
    scale=2,
    min_font_size=10
).generate(' '.join(cleaned_words))

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
